```python
# 조건문
# 점수를 입력받아 등급을 출력하세요
# 등급표 81~100	A / 61~80	B / 41~60	C / 21~40	D / 0~20	E

score = int(input("점수를 입력하세요: "))

if score >= 81: 
    print("점수는",score,"이며, A등급입니다.")
elif score >=61:
    print("점수는",score,"이며, B등급입니다.")
elif score >=41:
    print("점수는",score,"이며, C등급입니다.")

else:
    print("점수는",score,"이며 D등급 입니다. ")

```

```python
def pivo(n):
    a,b = 1,1
    if n==1 or n==2:
        return 1
        
    for i in range(1,n):
        a,b = b, a+b

    return a

print(pivo(10))
```

```python
# 클래스
# 사람 (Human) 클래스를 정의하고 (이름, 나이, 성별)을 받는 생성자를 추가하세요.
class Human:
    def __init__(self, name, age, sex):
        self.name = name
        self.age = age
        self.sex = sex

man = Human("dohyun", 25, "male")
print(man.name)
print(man.sex)
print(man.age)
```
